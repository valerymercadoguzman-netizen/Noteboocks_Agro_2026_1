## Ejemplos Sencillos de Python en Agroindustria

Aquí tienes tres scripts básicos que ilustran tareas comunes en el sector:

### 1. Cálculo del Índice de Madurez

Este código ayuda a determinar si un lote de frutas está listo para el mercado basándose en la relación sólidos solubles/acidez.

```
# Variables de entrada
grados_brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Fruta lista para exportación.")
else:
    print("Estado: Requiere más tiempo de maduración.")
```

### 2. Conversión de Unidades de Humedad

Ideal para laboratorios de control de calidad donde se recibe materia prima en base húmeda y se necesita pasar a base seca.

```
# Peso inicial y humedad medida
peso_total = 100  # kg
porcentaje_humedad = 20  # %

# Cálculo de materia seca
materia_seca = peso_total * (1 - porcentaje_humedad / 100)

print(f"De {peso_total}kg de producto, tienes {materia_seca}kg de materia seca real.")
```

### 3. Estimación de Capacidad de Almacenamiento

Un cálculo rápido para saber cuántos silos o contenedores necesitas para una cosecha de granos.

```
# Datos del silo (Cilíndrico)
radio = 3  # metros
altura = 10  # metros
densidad_grano = 750  # kg/m3 (ejemplo para maíz)

import math
volumen = math.pi * (radio**2) * altura
capacidad_total = volumen * densidad_grano

print(f"La capacidad estimada del silo es de {round(capacidad_total, 2)} kg de grano.")
```
